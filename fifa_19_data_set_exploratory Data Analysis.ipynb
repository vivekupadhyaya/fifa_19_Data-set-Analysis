# FIFA 19 Player Data Visualization

# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# READING FILE

```python
df = pd.read_csv('Fifa_19_dataset.csv')
```

```python
df.head()
```

# NUMBER OF ROWS AND COLUMNS ARE 

```python
df.shape  
```

```python
df.columns = [c.replace(' ', '') for c in df.columns]
```

```python
df.drop("Unnamed:0",axis=1,inplace=True)
df.head()
```

Some Analysis and Findings on Wage

```python
df["Wage"].head()
```

```python
df.Wage = df.Wage.str.replace("€","")
df.Wage = df.Wage.str.replace("K","").astype("float")
df.Wage.head()
```

unique players are in the data set

```python
df.ID.unique().shape
```

```python
df.info()
```

Preferred Foot, Overall Rating and Wage

```python
counts_preferred_foot = df["PreferredFoot"].value_counts()
counts_preferred_foot = counts_preferred_foot.reset_index()
counts_preferred_foot.columns = ["PreferredFoot","Count"]
print(counts_preferred_foot)
```

From analysis above, it has been found that 13948 playes use Right leg preferably and 4211 players use preferably Left leg. It can be observed from bar chart below

```python
pt = sns.catplot(y="Count",
            x="PreferredFoot", data=counts_preferred_foot,
           palette="colorblind",height=5, aspect=2, kind="bar");
plt.style.use("ggplot")
plt.title("Figure 1",
          fontsize=25);
```

Does preffered foot Vs overall ranking

```python
sns.catplot(x="PreferredFoot", y="Overall", data=df,
            aspect=2, kind="bar", palette="cubehelix");
plt.style.use("ggplot")
plt.title("Figure 2 ",
          fontsize=25);
```

From figure, It is not very clear that which preffered foot get more rating. So it requires some more statistical test analysis

```python
sns.catplot(x="PreferredFoot", y="Wage", data=df, aspect=2, kind="bar",palette="bright");
plt.style.use("ggplot")
plt.title("Figure 3",
          fontsize=25);
```

# Nationality, Overall rating and Wage

```python
counts_Nationality = df["Nationality"].value_counts()
counts_Nationality = counts_Nationality.reset_index()
counts_Nationality.columns= ["Nations","Counts"]
counts_Nationality.head()
```

```python
counts_Nationality_top20 = counts_Nationality.iloc[0:20,:]
print(counts_Nationality_top20)
```

```python
sns.catplot(y="Nations",
            x="Counts",data=counts_Nationality_top20,
            palette="colorblind",height=9,kind="bar",aspect=2)
plt.title("Figure : \n\n\n Nation Wise Players Counts for top 20 country",
          fontsize=25);
```

```python
avgwageoverall = df.groupby("Nationality", as_index=False)["Wage","Overall"].mean()
avgwageoverall.sort_values(by="Wage",inplace=True,ascending=False)
avgwageoverall.head()
```

```python
top10bywage = avgwageoverall.iloc[0:10,:]
top10bywage
```

```python
sns.catplot(y="Nationality",
            x="Wage",data=top10bywage,
            palette="colorblind",height=5,kind="bar",aspect=2)
plt.title("Figure 4",
          fontsize=25);
```

```python
avgwageoverall.sort_values(by="Overall",inplace=True,ascending=False)
top10byoverall = avgwageoverall.iloc[0:10,:]
top10byoverall
```

```python
sns.catplot(y="Nationality",
            x="Overall",data=top10byoverall,
           height=5,kind="bar",aspect=2,palette="colorblind")
plt.title("Figure 5:",
          fontsize=25);
```

* Findings Nationality and Overall Rating and Wage.
* In this dataset most of the players are from England which is followed by Germany.
* Nationality affect wage and overall rating of Player

# Does Age affect, Overall Rating and Wage ?

Age distribution of players

```python
x = df['Age']
plt.figure(figsize=(18,10))
ax = sns.countplot(x, color='g')
ax.set_xlabel(xlabel = 'Age of the Players', fontsize = 16)
ax.set_title(label = 'Distribution of Age of the Players', fontsize = 20)
plt.show()
```

```python
x = df['Age']
plt.figure(figsize = (12, 8))
plt.style.use('ggplot')
ax = sns.distplot(x, bins = 20, kde = True, color='g')
ax.set_xlabel(xlabel = 'Age of the Players', fontsize = 16)
ax.set_title(label = 'Histogram for Age distribution of Players', fontsize = 20)
plt.show()
```

```python
df.Age.value_counts()
```

```python
sns.lmplot(x="Age", y="Wage",data=df,markers="v",
            scatter_kws={"color": "blue"},
         line_kws={"linewidth":3,"color":"red"},aspect=2);
plt.title("Figure : \n\n\n Age vs Wage");
```

```python
sns.lmplot(x="Age", y="Overall",data=df,markers="o",
           order=3, ci=None, scatter_kws={"color": "red"},
         line_kws={"linewidth":2,"color":"green"},aspect=2);

plt.title("Figure : \n\n\n Overall Rating vs Wage");
```

# Findings Age and Overall Rating and Wage:
* Age lies between 11 to 95.
* Age of a Player does not affect much on their Wage.
* As Age increase the Overall rating increases.
